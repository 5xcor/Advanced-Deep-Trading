```python
def make_features_from_window(X_train_b, X_val, X_test, features):        
    X_train = np.array([x[features].fillna(0.).values.tolist() for x in X_train_b]) 
    X_val = np.array([x[features].fillna(0.).values.tolist() for x in X_val])
    X_test = np.array([x[features].fillna(0.).values.tolist() for x in X_test])
    return X_train, X_val, X_test
```

```python
def balance_XY_three_classes(X_train, Y_train):
    train_pos = [(x, y) for x, y in zip(X_train, Y_train) if y == 1]
    train_neg = [(x, y) for x, y in zip(X_train, Y_train) if y == -1]
    train_neu = [(x, y) for x, y in zip(X_train, Y_train) if y == 0]

    min_length = min(
        [len(train_pos), 
         len(train_neg), 
         len(train_neu)]
        )

    X_train_b, Y_train_b = [], []
    for x, y in train_pos[-min_length:]:
        X_train_b.append(x)
        Y_train_b.append(y)
    for x, y in train_neg[-min_length:]:
        X_train_b.append(x)
        Y_train_b.append(y)
    for x, y in train_neu[-min_length:]:
        X_train_b.append(x)
        Y_train_b.append(y)
        
    return X_train_b, Y_train_b
```

# Fixed horizon

```python
X, labels = [], []
for i in range(WINDOW, N_BARS, HORIZON):
    
    window = tick_bars.iloc[i-WINDOW:i]
#     window = tick_bars.iloc[i]
    now = tick_bars.close[i]
    future = tick_bars.close[i+HORIZON]
    ret = (future - now) / now
    
    X.append(window)
    if ret > T:
        labels.append(1)
    elif ret < -T:
        labels.append(-1)
    else:
        labels.append(0)
```

```python
plt.figure()
plt.hist(labels, bins = 5, alpha = 0.7)
plt.show()
```

```python
X_train, X_val, X_test = X[:int(len(X) * 0.5)], X[int(len(X) * 0.5):int(len(X) * 0.7)], X[int(len(X) * 0.7):]
Y_train, Y_val, Y_test = labels[:int(len(X) * 0.5)], labels[int(len(X) * 0.5):int(len(X) * 0.7)], labels[int(len(X) * 0.7):]

X_train, Y_train = balance_XY_three_classes(X_train, Y_train)
X_val, Y_val = balance_XY_three_classes(X_val, Y_val)

lbr = LabelBinarizer()
Y_train = lbr.fit_transform(Y_train)
Y_val = lbr.transform(Y_val)
Y_test = lbr.transform(Y_test)

X_train, X_val, X_test = make_features_from_window(X_train, X_val, X_test, FEATURES)
```

```python
def cnn(shape):
    main_input = Input(shape=shape, name='main_input')
    x = LSTM(10, recurrent_dropout=0.25)(main_input)
    output = Dense(3, activation = "softmax", kernel_regularizer=l2(1e-5))(x)
    
    final_model = Model(inputs=[main_input], outputs=[output])
    return final_model

model = cnn((WINDOW, len(X_train[0][0]), ))
model.summary()

model.compile(optimizer=Adam(lr = 0.001),  
                        loss=['categorical_crossentropy'], 
                        metrics = ['accuracy'])

checkpointer = ModelCheckpoint(filepath="test.hdf5", verbose=0, save_best_only=True)
es = EarlyStopping(monitor='val_loss', patience=5)

history = model.fit(X_train, Y_train, 
              epochs = 1000, 
              batch_size = 128, 
              verbose=True, 
              validation_data = (X_val, Y_val),
              callbacks=[checkpointer, es],
              shuffle=True)

model.load_weights("test.hdf5")
```

```python
pred = model.predict(X_test)

print classification_report([np.argmax(y) for y in Y_test],
                            [np.argmax(y) for y in pred])

print confusion_matrix([np.argmax(y) for y in Y_test],
                            [np.argmax(y) for y in pred])

plt.plot()
plt.hist([np.argmax(y) for y in Y_test], bins = 5, alpha = 0.5, label = 'Test data')
plt.hist([np.argmax(y) for y in pred], bins = 5, alpha = 0.5, label = 'Predicted data')
plt.legend()
plt.show()
```

# Volatility horizon

```python
X, labels = [], []
for i in range(WINDOW, N_BARS, HORIZON):
    window = tick_bars.iloc[i-WINDOW:i]
    now = tick_bars.close[i]
    future = tick_bars.close[i+HORIZON]
    ret = (future - now) / now
    
    window_abs_returns = np.abs(window.close.pct_change())
    Ti = np.std(window_abs_returns) + np.mean(window_abs_returns)
    
    X.append(window)
    if ret > Ti:
        labels.append(1)
    elif ret < -Ti:
        labels.append(-1)
    else:
        labels.append(0)
```

```python
plt.figure()
plt.hist(labels, bins = 5, alpha = 0.7)
plt.show()
```

```python
X_train, X_val, X_test = X[:int(len(X) * 0.5)], X[int(len(X) * 0.5):int(len(X) * 0.7)], X[int(len(X) * 0.7):]
Y_train, Y_val, Y_test = labels[:int(len(X) * 0.5)], labels[int(len(X) * 0.5):int(len(X) * 0.7)], labels[int(len(X) * 0.7):]

X_train, Y_train = balance_XY_three_classes(X_train, Y_train)
X_val, Y_val = balance_XY_three_classes(X_val, Y_val)

lbr = LabelBinarizer()
Y_train = lbr.fit_transform(Y_train)
Y_val = lbr.transform(Y_val)
Y_test = lbr.transform(Y_test)

X_train, X_val, X_test = make_features_from_window(X_train, X_val, X_test, FEATURES)
```

```python
def cnn(shape):
    main_input = Input(shape=shape, name='main_input')
    x = LSTM(10)(main_input)
    output = Dense(3, activation = "softmax", kernel_regularizer=l2(1e-6))(x)
    
    final_model = Model(inputs=[main_input], outputs=[output])
    return final_model

model = cnn((WINDOW, len(X_train[0][0]), ))
model.summary()

model.compile(optimizer=Adam(lr = 0.001),  
                        loss=['categorical_crossentropy'], 
                        metrics = ['accuracy'])

checkpointer = ModelCheckpoint(filepath="test.hdf5", verbose=0, save_best_only=True)
es = EarlyStopping(monitor='val_loss', patience=5)

history = model.fit(X_train, Y_train, 
              epochs = 1000, 
              batch_size = 128, 
              verbose=True, 
              validation_data = (X_val, Y_val),
              callbacks=[checkpointer, es],
              shuffle=True)

model.load_weights("test.hdf5")
```

# Meta labeling

```python
def get_meta_barier(future_window, last_close, min_ret, tp, sl, vertical_zero = False):
    '''
        XXX
    '''
    if vertical_zero:
        min_ret_situation = [0, 0, 0]
    else:
        min_ret_situation = [0, 0]
        
        
    differences = np.array([(fc - last_close) / last_close for fc in future_window])
    
    # Are there gonna be fluctuations within min_ret???
    min_ret_ups = np.where((differences >= min_ret) == True)[0]
    min_ret_downs = np.where((differences < -min_ret) == True)[0]
  
    if (len(min_ret_ups) == 0) and (len(min_ret_downs) == 0):
        if vertical_zero:
            min_ret_situation[2] = 1
        else:
            if differences[-1] > 0:
                min_ret_situation[0] = 1
            else:
                min_ret_situation[1] = 1            
    else:
        if len(min_ret_ups) == 0: min_ret_ups = [np.inf]
        if len(min_ret_downs) == 0: min_ret_downs = [np.inf]
        if min_ret_ups[0] > min_ret_downs[0]:
            min_ret_situation[0] = 1
        else:
            min_ret_situation[1] = 1
        
    #  Take profit and stop losses indices
    take_profit = np.where((differences >= tp) == True)[0]
    stop_loss = np.where((differences < sl) == True)[0]
    
    # Fluctuation directions coincide with take profit / stop loss actions?
    if min_ret_situation[0] == 1 and len(take_profit) != 0:
        take_action = 1
    elif min_ret_situation[1] == 1 and len(stop_loss) != 0:
        take_action = 1
    else:
        take_action = 0.
    
    return min_ret_situation, take_action
```

```python
TP = T
SL = -T

X, Y, Y2 = [], [], []
for i in range(WINDOW, N_BARS, HORIZON):
    window = tick_bars.iloc[i-WINDOW:i]
    now = tick_bars.close[i]
    future_window = tick_bars.close[i:i+HORIZON]
    
    window_abs_returns = np.abs(window.close.pct_change())
    Ti = np.std(window_abs_returns) + np.mean(window_abs_returns)
    
    min_ret_situation, take_action = get_meta_barier(future_window, now, Ti, TP, SL, False)
    X.append(window)
    Y.append(np.argmax(min_ret_situation))
    Y2.append(take_action)
```

```python
plt.figure()
plt.hist(Y, bins = 5, alpha = 0.7)
plt.show()
```

```python
X_train, X_val, X_test = X[:int(len(X) * 0.5)], X[int(len(X) * 0.5):int(len(X) * 0.7)], X[int(len(X) * 0.7):]
Y_train, Y_val, Y_test = Y[:int(len(X) * 0.5)], Y[int(len(X) * 0.5):int(len(X) * 0.7)], Y[int(len(X) * 0.7):]

X_train, Y_train = balance_XY_three_classes(X_train, Y_train)
X_val, Y_val = balance_XY_three_classes(X_val, Y_val)

X_train, X_val, X_test = make_features_from_window(X_train, X_val, X_test, FEATURES)
```

```python
def cnn(shape):
    main_input = Input(shape=shape, name='main_input')
#     x = Flatten()(main_input)
    x = LSTM(10)(main_input)
    output = Dense(1, activation = "sigmoid", kernel_regularizer=l2(1e-6))(x)
    
    final_model = Model(inputs=[main_input], outputs=[output])
    return final_model

model = cnn((WINDOW, len(X_train[0][0]), ))
model.summary()

model.compile(optimizer=Adam(lr = 0.001),  
                        loss=['binary_crossentropy'], 
                        metrics = ['accuracy'])

checkpointer = ModelCheckpoint(filepath="test.hdf5", verbose=0, save_best_only=True)
es = EarlyStopping(monitor='val_loss', patience=5)

history = model.fit(X_train, Y_train, 
              epochs = 500, 
              batch_size = 128, 
              verbose=True, 
              validation_data = (X_val, Y_val),
              callbacks=[checkpointer, es],
              shuffle=True)

model.load_weights("test.hdf5")
```

```python
pred = model.predict(X_test)

print classification_report(Y_test, np.rint(pred))
print confusion_matrix(Y_test, np.rint(pred))

plt.plot()
plt.hist(Y_test, bins = 3, alpha = 0.5, label = 'Test data')
plt.hist(np.rint(pred), bins = 3, alpha = 0.5, label = 'Predicted data')
plt.legend()
plt.show()
```

```python
def cnn(shape):
    main_input = Input(shape=shape, name='main_input')
    aux_input = Input((1, ), name='meta')
    
    x = LSTM(10)(main_input)
#     x = Flatten()(main_input)
    x = concatenate([x, aux_input])
    output = Dense(1, activation = "sigmoid", kernel_regularizer=l2(1e-6))(x)
    
    final_model = Model(inputs=[main_input, aux_input], outputs=[output])
    return final_model

model = cnn((WINDOW, len(X_train[0][0]), ))
model.summary()

model.compile(optimizer=Adam(lr = 0.001),  
                        loss=['binary_crossentropy'], 
                        metrics = ['accuracy'])

checkpointer = ModelCheckpoint(filepath="test.hdf5", verbose=0, save_best_only=True)
es = EarlyStopping(monitor='val_loss', patience=5)

history = model.fit([X_train, P_train], Y_train, 
              epochs = 500, 
              batch_size = 128, 
              verbose=True, 
              validation_data = ([X_test, P_test], Y_test),
              callbacks=[checkpointer, es],
              shuffle=True)

model.load_weights("test.hdf5")
```
